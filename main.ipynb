```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import sys

inImage = cv2.imread("./data/test.jpg")
inImage = np.float32(inImage) / 255.0

# Calculate Gradients
gradX = cv2.Sobel(inImage, cv2.CV_32F, 1, 0, ksize=1)
gradY = cv2.Sobel(inImage, cv2.CV_32F, 0, 1, ksize=1)

mag, angle = cv2.cartToPolar(gradX, gradY, angleInDegrees = True)

plt.imshow(inImage)
plt.show()

plt.imshow(mag)
plt.show()

plt.imshow(angle)
plt.show()
```
